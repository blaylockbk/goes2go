# GOES-19 (GOES-East)

```python
from goes2go import GOES

import matplotlib.pyplot as plt
```

```python
G = GOES(satellite=19, product="ABI-L2-MCMIP", domain="C")
ds = G.latest()
ds
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(ds["CMI_C07"].data, cmap="Greys")
plt.title("GOES-19; ABI Channel 7", loc="left")
plt.title(ds.t.dt.strftime("%H:%M UTC %b %d, %Y").item(), loc="right")
plt.colorbar(label="Brightness Temperature [K]", pad=0.01, shrink=0.5)
```

> When calling get_latest, the fs.ls command returns a cached list of the files on s3. Adding refresh=True (or having kwargs to set this) to the following should allow an updates enumeration of files on s3:

```python
import s3fs
import pandas as pd
from datetime import datetime, timedelta

from goes2go.data import _goes_file_df, goes_latest, _product
```

```python
start = datetime.utcnow() - timedelta(hours=1)
end = datetime.utcnow()
satellite = "noaa-goes16"
product = "ABI-L2-MCMIPM"
_goes_file_df(satellite, product, start, end)
```

```python
start = datetime.utcnow() - timedelta(hours=1)
end = datetime.utcnow()
_goes_file_df(satellite, product, start, end, refresh=False)
```

```python
start = datetime.utcnow() - timedelta(hours=1)
end = datetime.utcnow()
_goes_file_df(satellite, product, start, end, refresh=True)
```
